# Array Operations

In machine learning, data rarely comes in the correct shape.

We often need to reshape, convert, and manage memory efficiently.

```python
import numpy as np
```

## Reshaping Arrays

Reshape changes dimensions without changing data.

```python
arr = np.arange(1, 13)
print(arr)

reshaped = arr.reshape(3,4)
print(reshaped)
```

## Flattening Arrays

Convert multi-dimensional array → 1D array

```python
flat = reshaped.flatten()
print(flat)
```

## NumPy Data Types

NumPy supports efficient numeric types

```python
arr = np.array([1,2,3], dtype=np.int32)
print(arr.dtype)
```

## astype()

Convert array data type

```python
float_arr = arr.astype(np.float64)
print(float_arr)
print(float_arr.dtype)
```

## Copy vs View

View → shares memory  
Copy → new memory

```python
a = np.array([1,2,3,4])
b = a.view()

b[0] = 99

print("a:", a)
print("b:", b)
```

```python
c = a.copy()
c[1] = 100

print("a:", a)
print("c:", c)
```
