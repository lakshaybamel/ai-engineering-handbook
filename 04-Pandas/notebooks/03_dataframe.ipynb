# Pandas DataFrame

A DataFrame is a two-dimensional labeled data structure.

It contains:<br>
rows + columns

Each column is a Pandas Series.

```python
import pandas as pd
```

## Creating DataFrame from Dictionary

```python
data = {
    "Name": ["Lakshay", "Amit", "Harsh"],
    "Age": [23, 25, 22],
    "City": ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)
df
```

## Columns & Index

```python
df.columns
df.index
```

## Selecting a Column (Series)

```python
df["Name"]
```

## Selecting Multiple Columns

```python
df[["Name", "City"]]
```

## Adding New Column

```python
df["Salary"] = [50000, 60000, 55000]
df
```

## Deleting Column

```python
df.drop("Salary", axis=1)
```

## Selecting Rows using loc

```python
df.loc[0]
df.loc[1]
```

## Selecting Rows using iloc

```python
df.iloc[0]
df.iloc[1]
```

## Basic Info

```python
df.info()
df.describe()
```

## Conclusion

DataFrame is the main structure used for dataset manipulation.
