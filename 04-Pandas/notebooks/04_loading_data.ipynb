# Loading Data in Pandas

In real projects, data comes from files.

Common formats:<br>
• CSV (most common)<br>
• JSON

```python
import pandas as pd
```

## Loading CSV File

```python
sales = pd.read_csv("../data/raw/sales.csv")
sales.head()
```

## Loading Another CSV

```python
customers = pd.read_csv("../data/raw/customers.csv")
customers.head()
```

## Loading Large Dataset (Air Quality)

```python
air = pd.read_csv("../data/raw/air_quality.csv")
air.head()
```

## Loading JSON File

```python
json_data = pd.read_json("../data/raw/sample.json")
json_data
```

## Inspect Data Shapes

```python
sales.shape, customers.shape, air.shape, json_data.shape
```

## Conclusion

Pandas can load structured data from multiple formats.
