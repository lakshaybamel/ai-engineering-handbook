# Pandas Series

A Series is a one-dimensional labeled array.

It can store:<br>
numbers, strings, booleans, etc.

Think of Series as:<br>
a single column of a table.

```python
import pandas as pd
```

## Creating Series from List

```python
data = [10,20,30,40]
s = pd.Series(data)

s
```

Each value has an index automatically assigned.

```python
s.index
s.values
```

## Custom Index

```python
s = pd.Series([100,200,300], index=["a","b","c"])
s
```

## Accessing Elements

```python
s["a"]
s["c"]
```

## Mathematical Operations

```python
s + 10
s * 2
```

## Series from Dictionary

```python
data = {"Delhi":40, "Mumbai":35, "Chennai":38}
temp = pd.Series(data)

temp
```

## Handling Missing Values

```python
temp.isnull()
temp.notnull()
```

## Basic Statistics

```python
temp.mean()
temp.max()
temp.min()
```

## Conclusion

Series is the building block of DataFrame.
